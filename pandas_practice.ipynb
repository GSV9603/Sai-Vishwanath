```python
data={"weight":[55,65,75,80,90,95], "height":[165,172,183,146,166,169]}
bmi=pd.DataFrame(data)
bmi
```

```python
bmi["BMI"]=bmi["weight"]/(bmi["height"]/100)**2
bmi
```

```python
bmi.loc[3]=[73,np.nan,np.nan]
bmi
```

```python
bmi.loc[5]=[np.nan,65,np.nan]
bmi
```

```python
bmi.isnull().sum()
```

```python
bmi["weight"]=bmi["weight"].fillna(77)
bmi
```

```python
bmi["height"]=bmi["height"].fillna(177)
bmi
```

```python
bmi["BMI_New"]=bmi["weight"]/(bmi["height"]/100)**2
bmi
```

```python
bmi.drop("BMI",axis=1, inplace=True)
```
